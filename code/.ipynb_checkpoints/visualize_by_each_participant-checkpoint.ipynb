```python
# visualize by each partipant to check the crop result
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dir = "./../result/"
participant1 = pd.read_csv(dir + "1.csv")
participant1_seg = pd.read_csv(dir + "1_final.csv")
```

```python
def draw_plot(data, title):
    x = data['accelX']
    y = data['accelY']
    z = data['accelZ']
    
    index = [i for i in range(len(x))]
    
    plt.plot(index, x, 'r-', label='X')
    plt.plot(index, y, 'b-', label='Y')
    plt.plot(index, z, 'g-', label='Z')

    plt.title(title)
    plt.xlabel('index')
    plt.ylabel('Accel')
    
    plt.legend()
    plt.show()
```

```python
activities = np.unique(participant1.ActivityNumber)


for activity in activities:
    data = participant1.loc[participant1.ActivityNumber == activity,]
    draw_plot(data, "Accelerometer data for Activity #%d"%activity)
    
    data_sep = participant1_seg.loc[participant1_seg.ActivityNumber == activity,]
    index = np.unique(data_sep.Index)

    for i in index:
        data_show = participant1_seg.query("Index == %d & ActivityNumber == %d"%(i, activity))
        draw_plot(data_show, "Accelerometer data for Activity # %d, [%d-%d]"%(activity, i*300, (i+1)*300))
    
```
